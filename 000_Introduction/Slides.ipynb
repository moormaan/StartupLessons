# Why now?

The first working electrostatic telegraph was built by the English inventor Francis Ronalds. In **1816** he laid down eight miles of wire insulated in glass tubing in his garden and connected both ends to two clocks marked with the letters of the alphabet. Electrical impulses sent along the wire were used to transmit messages.
