<img src="http://electronics360.globalspec.com/images/assets/241/5241/Date_of_introduction-Fullsize.jpg" alt="Moore's Law" style="height: 600px;"/>

```python
x = np.linspace(0, 10, 100)
y = np.exp2(x/2)

plt.figure()
plt.plot(x, y)
plt.xlabel('$x$')
plt.ylabel('$2^{(x/2)}$')
plt.gcf().set_size_inches(w_in, h_in, forward=True)
```
